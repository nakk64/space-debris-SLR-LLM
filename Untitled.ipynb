```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import subprocess

# --- 1. Set Up Paths ---
DATA_DIR = "./envisat"  # Folder containing year-based subfolders
OUTPUT_DIR = "./processed"
MODEL_PATH = "/home/inet/llama.cpp/models/tinyllama-1.1b-chat-v1.0.Q4_K_M.gguf"
LLAMA_CLI = "/home/inet/llama.cpp/build/bin/llama-cli"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- 2. Enhanced CRD Parser ---
def parse_crd(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith('10') or line.startswith('01'):  # Data lines usually start with '10' or '01'
                parts = line.strip().split()
                try:
                    ts = float(parts[1])  # Timestamp (TDT seconds or MJD)
                    rng = float(parts[2])  # Range in light travel time (seconds)
                    station = parts[3] if len(parts) > 3 else None
                    sat_id = parts[4] if len(parts) > 4 else None
                    az = float(parts[10]) if len(parts) > 10 and parts[10].replace('.', '', 1).isdigit() else None
                    el = float(parts[11]) if len(parts) > 11 and parts[11].replace('.', '', 1).isdigit() else None
                    signal_strength = float(parts[13]) if len(parts) > 13 and parts[13].isdigit() else None

                    data.append({
                        'timestamp': ts,
                        'range': rng,
                        'station': station,
                        'sat_id': sat_id,
                        'azimuth': az,
                        'elevation': el,
                        'signal_strength': signal_strength
                    })
                except Exception as e:
                    continue  # Skip malformed lines
    return pd.DataFrame(data)

# --- 3. Recursive File Processor ---
def process_all_files(root_dir):
    all_data = pd.DataFrame()
    for root, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".fr2"):
                full_path = os.path.join(root, file)
                print(f"Parsing {full_path}")
                df = parse_crd(full_path)
                if not df.empty:
                    all_data = pd.concat([all_data, df], ignore_index=True)
                    df.to_csv(os.path.join(OUTPUT_DIR, f"{file}.csv"), index=False)
    return all_data

# --- 4. Run It ---
print("🔄 Reading and parsing all CRD .fr2 files...")
df_all = process_all_files(DATA_DIR)

if not df_all.empty:
    df_all.to_csv(os.path.join(OUTPUT_DIR, "envisat_all_parsed.csv"), index=False)
    print(f"✅ Saved combined dataset to {os.path.join(OUTPUT_DIR, 'envisat_all_parsed.csv')}")

    # --- 5. Anomaly Detection ---
    median_range = df_all['range'].median()
    std_range = df_all['range'].std()
    df_all['anomaly'] = np.abs(df_all['range'] - median_range) > 3 * std_range
    anomalies = df_all[df_all['anomaly']]
    print(f"\n🔍 Found {len(anomalies)} anomalies (>3σ):")
    print(anomalies[['timestamp', 'range']].head())

    # --- 6. Plotting ---
    plt.figure(figsize=(12, 6))
    plt.plot(df_all['timestamp'], df_all['range'], marker='o', linestyle='', markersize=2, label='Range')
    if not anomalies.empty:
        plt.scatter(anomalies['timestamp'], anomalies['range'], color='red', label='Anomalies', zorder=5)
    plt.title("Envisat Laser Range Over Time")
    plt.xlabel("Timestamp (MJD / TDT)")
    plt.ylabel("Range (light travel time in s)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "envisat_range_plot.png"))
    plt.show()

    # --- 7. Prepare Prompt for LLaMA ---
    sample_data = df_all[['timestamp', 'range']].head(5).to_string(index=False)
    total_obs = len(df_all)
    most_active_station = df_all['station'].mode()[0] if not df_all['station'].isna().all() else "unknown"

    prompt = f"""
You are an expert in space situational awareness. Analyze the following Envisat laser ranging data.

1. Identify and explain any range anomalies (greater than 3σ from the median).
2. Estimate if there's any orbital decay or pattern.
3. Suggest recommendations for next tracking steps.

Summary:
- Total observations: {total_obs}
- Median range: {median_range:.6f} s
- Standard deviation: {std_range:.6f} s
- Most active station: {most_active_station}

Data sample (first 5 entries):
{sample_data}
"""

    print("\n🧠 Sending prompt to LLaMA model...\n")
    print(prompt)

    # --- 8. Run LLaMA Model ---
    llama_cmd = [
        LLAMA_CLI,
        "-m", MODEL_PATH,
        "-p", prompt,
        "--ctx-size", "4096",
        "--n-predict", "512"
    ]

    result = subprocess.run(llama_cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    response = result.stdout.strip()

    print("\n🤖 LLaMA Output:\n")
    print(response)

    # --- 9. Save Report ---
    report = f"""
# Envisat SLR Tracking Report

## Summary
- Total observations: {total_obs}
- Median range: {median_range:.6f} s
- Standard deviation: {std_range:.6f} s
- Most active station: {most_active_station}

## Sample Data
{sample_data}

## LLaMA Analysis
{response}
"""

    report_path = os.path.join(OUTPUT_DIR, "envisat_analysis_report.md")
    with open(report_path, "w") as f:
        f.write(report)
    print(f"\n📄 Report saved to {report_path}")

else:
    print("❌ No valid data found in the CRD files.")
```
